```python
cols = ['Location', 'WindGustDir', 'WindDir9am', 'WindDir3pm',
       'RainToday', 'RainTomorrow']
two_values = []

for i in cols:
        if len(df[i].unique()) <= 2:
            two_values.append(i)
        
            
print(two_values)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for i in two_values:
    df_new[i] = le.fit_transform(df_new[i])

print(df_new["RainToday"].unique())
print(df_new["RainTomorrow"].unique())
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

ct = ColumnTransformer([('encoder', OneHotEncoder(), [0,6,8,9])], remainder='passthrough')
x = ct.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2)
```

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler(with_mean = False)
x_train = scale.fit_transform(x_train)
x_test = scale.transform(x_test)
```

# Model building

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
```

```python
model.add(Dense(units = 100,activation = "relu"))
```

```python
model.add(Dense(units = 100,activation = "relu"))
```

```python
model.add(Dense(units = 2, activation = "sigmoid"))
```

```python
model.compile(optimizer = "adam",loss = "binary_crossentropy")
```

```python
model.fit(x_train,y_train,epochs = 50,batch_size = 32)
```

```python
predicted = model.predict(x_test)
predicted = predicted > 0.5
predicted
```

```python
y_test
```

```python
np.concatenate((y_test,predicted),1)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
accuracy_score(predicted,y_test)
```

```python
confusion_matrix(predicted.argmax(axis = 1),y_test.argmax(axis = 1))
```
